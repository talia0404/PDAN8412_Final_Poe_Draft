```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

Define Paths

```python
train_dir = "/content/dataset/GemStones/GemStones/GemStonesTensorflow/train"
val_dir = "/content/dataset/GemStones/GemStonesTensorflow/valid"
test_dir = "/content/dataset/GemStones/GemStones/GemStonesTensorflow/test"
```

Set Up ImageDataGenerator for Data Loading and Augmentation

```python
# Define training data generator with augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    horizontal_flip=True,
    rotation_range=15,
    shear_range=0.1,
    brightness_range=[0.9, 1.1]
)
```

```python
# Validation data generator (no augmentation)
val_datagen = ImageDataGenerator(rescale=1.0/255)
```

```python
# Load images in batches from directory
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(640, 640),
    batch_size=32,
    class_mode='categorical'
)
```
